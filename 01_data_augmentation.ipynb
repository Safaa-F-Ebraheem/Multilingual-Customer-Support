```python
import pandas as pd
import nlpaug.augmenter.word as naw
import nlpaug.augmenter.sentence as nas
```

### Load Raw Data

```python
df_raw = pd.read_csv('reviews.csv')
```

```python


# Initialize the augmenters
synonym_aug = naw.SynonymAug(aug_p=0.1)  # Synonym replacement with a probability of 10%
contextual_word_aug = naw.ContextualWordEmbsAug(model_path='distilbert-base-uncased', aug_p=0.1)

# Function to augment text
def augment_text(text, augmenter):
    return augmenter.augment(text)

# Augment 'title' and 'content' columns
df_raw['augmented_title'] = df_raw['title'].apply(lambda x: augment_text(x, synonym_aug))
df_raw['augmented_content'] = df_raw['content'].apply(lambda x: augment_text(x, contextual_word_aug))

```

```python
df_raw.to_csv('reviews_augmented.csv', index=False)

```
